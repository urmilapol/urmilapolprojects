```python
from pycaret.datasets import get_data
diabetes = get_data('diabetes')
```

```python
# import the classification module 
from pycaret import classification
# setup the environment 
classification_setup = classification.setup(data= diabetes, target='Class variable')
```

```python
classification.compare_models()
```

```python
classification_catb = classification.create_model('xgboost')
```

```python
tuned_xgboost = classification.tune_model('xgboost')
```

```python
# create a model
xgboost = classification.create_model('xgboost')
# AUC plot
classification.plot_model(xgboost, plot = 'auc')
```

```python
# Decision Boundary
classification.plot_model(xgboost, plot = 'boundary')
```

```python
# Precision Recall Curve
classification.plot_model(xgboost, plot = 'pr')
```

```python
# Validation Curve
classification.plot_model(xgboost, plot = 'vc')
```

```python
classification.evaluate_model(xgboost)
```

```python
# create a model
xgboost = classification.create_model('xgboost')
# summary plot
classification.interpret_model(xgboost)
# correlation plot
classification.interpret_model(xgboost, plot = 'correlation')
```

```python
# create a model
xgboost = classification.create_model('xgboost')
# predict test / hold-out dataset
xgboost_holdout_pred = classification.predict_model(xgboost)
```

```python
predictions = classification.predict_model(xgboost, data = diabetes)
```
