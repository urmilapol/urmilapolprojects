```python
# ==========================================
# PART 1: STRINGS & NUMBERS
# ==========================================
print("--- Strings & Numbers ---")
phrase = "Draft Academy"
print(phrase.upper())             # Transforms to uppercase [00:31:12]
print(len(phrase))                # Gets length [00:32:43]
print(phrase[0])                  # Indexing: Starts at 0 [00:34:01]

num = -5
print(abs(num))                   # Absolute value [00:43:38]
print(pow(3, 2))                  # 3 squared = 9 [00:44:09]

# ==========================================
# PART 2: LISTS & TUPLES
# ==========================================
print("\n--- Lists & Tuples ---")
friends = ["Kevin", "Karen", "Jim"]
friends[1] = "Mike"               # Lists are mutable (can change) [01:09:47]
print(friends)

coordinates = (4, 5)              # Tuples are immutable (CANNOT change) [01:19:07]
# coordinates[1] = 10             # This would throw an error!

# ==========================================
# PART 3: FUNCTIONS
# ==========================================
print("\n--- Functions ---")
def say_hi(name, age):            # Defining a function with parameters [01:25:22]
    print("Hello " + name + ", you are " + str(age))

say_hi("Mike", 35)

def cube(num):                    # Return statements [01:34:16]
    return num * num * num

result = cube(4)
print(f"4 cubed is: {result}")

# ==========================================
# PART 4: LOGIC & LOOPS
# ==========================================
print("\n--- Logic & Loops ---")
is_tall = True
if is_tall:                       # If statement [01:40:09]
    print("You are tall")
else:
    print("You are short")

# For loop through an array [02:32:47]
for friend in friends:
    print("Friend: " + friend)

# ==========================================
# PART 5: CLASSES & OBJECTS
# ==========================================
print("\n--- Classes & Objects ---")
class Student:                    # The Blueprint [03:43:58]
    def __init__(self, name, major, gpa):
        self.name = name
        self.major = major
        self.gpa = gpa

    def on_honor_roll(self):      # Class Function [04:08:36]
        if self.gpa >= 3.5:
            return True
        return False

student1 = Student("Jim", "Business", 3.1)
student2 = Student("Pam", "Art", 3.8)

print(f"{student1.name} Honor Roll: {student1.on_honor_roll()}")
print(f"{student2.name} Honor Roll: {student2.on_honor_roll()}")
```

2. Core Data Structures [15:45]
DataFrame: A two-dimensional, tabular data structure with rows and columns (like an Excel sheet). Created using pd.DataFrame().

Series: A one-dimensional array-like object with labels. Created using pd.Series().

Indexing: Use .loc for label-based indexing and .iloc for integer-position-based indexing.

3. Data Manipulation & Cleaning [03:00:53]
Adding/Deleting: Use .insert() or .assign() to add columns. Use .drop() with axis=0 for rows or axis=1 for columns to delete data.

Cleaning: * Find nulls with .isnull().

Remove rows with nulls using .dropna().

Fill missing values with .fillna(value).

Handle duplicates with .duplicated() and .drop_duplicates().

4. Advanced Operations [05:12:13]
Grouping: The .groupby() method allows you to split data into groups based on criteria, apply functions, and combine results.

Statistics: Built-in methods like .sum(), .mean(), .median(), .max(), and .min() provide quick insights. Use .describe() for a full statistical summary.

Visualization: Pandas integrates with Matplotlib. Use the .plot() method with the kind parameter for histograms, pie charts, and scatter plots.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# PART 1: DATAFRAMES & SERIES
# ==========================================

# Creating a Series [01:15:56]
data_series = [10, 20, 30, 40, 50]
s = pd.Series(data_series, index=['a', 'b', 'c', 'd', 'e'])
print("Series:\n", s)

# Creating a DataFrame [00:19:45]
student_data = {
    'Name': ['Amit', 'John', 'Jacob', 'David', 'Steve'],
    'Rank': [1, 4, 3, 5, 2],
    'Marks': [95, 70, 80, 60, 90]
}
df = pd.DataFrame(student_data)
print("\nDataFrame:\n", df)

# ==========================================
# PART 2: INDEXING & SELECTION
# ==========================================

# Using .loc (Label-based) [00:23:05]
print("\nAmit's Marks:", df.loc[0, 'Marks'])

# Using .iloc (Position-based) [00:26:08]
print("\nFirst row via iloc:\n", df.iloc[0])

# ==========================================
# PART 3: DATA CLEANING [03:51:18]
# ==========================================

# Adding a null value for demonstration
df.loc[5] = ['Ben', 6, np.nan]

# Checking for Nulls
print("\nNull Values:\n", df.isnull())

# Filling Nulls
df_filled = df.fillna(0)
print("\nDataFrame Filled:\n", df_filled)

# ==========================================
# PART 4: STATISTICS & GROUPING [05:12:13]
# ==========================================

# Statistical Summary
print("\nStatistical Summary:\n", df_filled.describe())

# Grouping [04:53:21]
# Assume a 'Year' column was added
df_filled['Year'] = [2021, 2021, 2022, 2022, 2023, 2023]
grouped = df_filled.groupby('Year')['Marks'].mean()
print("\nAverage Marks per Year:\n", grouped)

# ==========================================
# PART 5: VISUALIZATION [05:33:32]
# ==========================================

# Simple Plot
df_filled.plot(x='Name', y='Marks', kind='bar')
plt.title('Student Marks Comparison')
plt.show()

# Pie Chart [05:43:55]
plt.pie(df_filled['Marks'], labels=df_filled['Name'], autopct='%1.1f%%')
plt.title('Marks Distribution')
plt.show()
```
