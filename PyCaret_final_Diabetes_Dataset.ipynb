```python
# import the classification module 
from pycaret.classification import *
# setup the environment 
classification_setup = classification.setup(data= diabetes, target='Class variable')
```

```python
classification.compare_models()
```

```python
classification_catb = classification.create_model('gbc')
```

```python
# create a model
gbc = classification.create_model('gbc')
# AUC plot
classification.plot_model(gbc, plot = 'auc')
```

```python
# Decision Boundary
classification.plot_model(gbc, plot = 'boundary')
```

```python
# Precision Recall Curve
classification.plot_model(gbc, plot = 'pr')
```

```python
# Validation Curve
classification.plot_model(gbc, plot = 'vc')
```

```python
classification.evaluate_model(gbc)
```
