# Python Examples for Different Types of Data Analysis

**1. Descriptive Analysis**

```python
import pandas as pd

# Sample dataset (sales data)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 240, 180, 300, 350]
}

df = pd.DataFrame(data)

# Descriptive statistics
print(df['Sales'].describe())  # Mean, median, min, max, etc.
print("\nMean Sales:", df['Sales'].mean())
print("Median Sales:", df['Sales'].median())
print("Standard Deviation:", df['Sales'].std())
```

**2. Diagnostic Analysis**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (sales vs. advertising spend)
data = {
    'Sales': [200, 240, 180, 300, 350],
    'Ad_Spend': [50, 60, 40, 70, 80]
}

df = pd.DataFrame(data)

# Correlation matrix
correlation = df.corr()
print(correlation)

# Heatmap visualization
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Between Sales & Ad Spend")
plt.show()
```

**3. Predictive Analysis**
Forecasting future trends using machine learning.
Example: Linear Regression for Sales Prediction


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data (Ad Spend vs Sales)
X = np.array([50, 60, 40, 70, 80]).reshape(-1, 1)  # Ad Spend
y = np.array([200, 240, 180, 300, 350])            # Sales

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict sales for $90 ad spend
prediction = model.predict([[90]])
print("Predicted Sales for $90 Ad Spend:", prediction[0])
```

4. Prescriptive Analysis
Recommending actions based on predictions.

Example: Optimizing Ad Budget for Max Profit

```python
# Assume:
# - Profit = Sales * 0.3 - Ad_Spend (30% margin)
# - We predict sales using regression model

def calculate_profit(ad_spend):
    predicted_sales = model.predict([[ad_spend]])[0]
    profit = predicted_sales * 0.3 - ad_spend
    return profit

# Test different ad spends
ad_spends = [50, 60, 70, 80, 90]
profits = [calculate_profit(ads) for ads in ad_spends]

optimal_ad_spend = ad_spends[np.argmax(profits)]
print("Optimal Ad Spend for Max Profit:", optimal_ad_spend)
```

5. Exploratory Data Analysis (EDA)
Visualizing patterns and relationships in data.

Example: EDA on Iris Dataset

```python
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset('iris')

# Pairplot to see relationships
sns.pairplot(iris, hue='species')
plt.show()

# Boxplot for feature distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, orient='h')
plt.title("Feature Distribution in Iris Dataset")
plt.show()
```

Complete Exploratory Data Analysis (EDA) Example in Python (Without Machine Learning)
We'll use the Titanic dataset (a classic dataset for EDA) to explore passenger demographics, survival rates, and relationships between variables.
1. Setup & Data Loading

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Basic dataset info
print("\nDataset Info:")
print(df.info())
```

2. Handling Missing Data

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Visualize missing data
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

3. Descriptive Statistics

```python
# Numerical columns summary
print("\nNumerical Summary:")
print(df.describe())

# Categorical columns summary
print("\nCategorical Summary:")
print(df.describe(include=['object', 'category', 'bool']))
```

4. Univariate Analysis (Single Variable Exploration)
Age Distribution
(Histogram shows most passengers were between 20-40 years old.)*


```python
plt.figure(figsize=(10, 5))
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Survival Rate
(Bar plot shows ~38% survived, ~62% died.)

```python
plt.figure(figsize=(6, 4))
df['survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.xticks(rotation=0)
plt.show()
```

5. Bivariate Analysis (Two Variables)
Survival by Passenger Class
(Higher-class passengers had better survival rates.)

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='pclass', hue='survived')
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class (1 = Highest, 3 = Lowest)")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

Fare vs. Survival
(Survivors generally paid higher fares.)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='survived', y='fare')
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
```

6. Multivariate Analysis (Multiple Variables)
Age vs. Fare Colored by Survival
(Younger passengers with higher fares had better survival chances.)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', alpha=0.6)
plt.title("Age vs. Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

Correlation Heatmap
(Shows relationships between numerical variables.)

```python
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
