```python
# ===============================================
# COMPLETE PYTHON ETL DEMO FROM MAVEN ANALYTICS VIDEO
# ===============================================
# Colab-ready: Extracts LA retail sales from CSVs/JSON/Parquet/DB/API,
# transforms (clean/merge/enrich), loads to SQLite.
#
# STEPS TO RUN:
# 1. Download dataset: https://maven-datasets.s3.us-east-1.amazonaws.com/YouTube+Downloads/Spring+Skillup.zip
# 2. Unzip to /content/spring_skillup/ (upload via Colab files panel)
# 3. Get FREE OpenWeather API key: https://openweathermap.orgfapi/api (Historical needs paid plan)
# 4. Run all cells sequentially.
# ===============================================

# ===== 0. SETUP & IMPORTS =====
import pandas as pd
import sqlite3
import glob
import requests
import time
from datetime import datetime
import seaborn as sns  # Pretty plots
import matplotlib.pyplot as plt

# Create SQLite DB
con = sqlite3.connect('/content/etl_demo.db')
print("âœ… SQLite DB connected: /content/etl_demo.db")

# ===== 1. EXTRACT: SINGLE CSV =====
print("\nðŸ“¥ 1. EXTRACT SINGLE CSV")
sales_df = pd.read_csv('/content/sample_data/LA_Retail_Sales.csv')
print(f"Shape: {sales_df.shape}")
display(sales_df.head())
sales_df.info()

# ===== 2. EXTRACT: MULTI-CSV (GLOB) =====
print("\nðŸ“¥ 2. EXTRACT MULTIPLE CSVs (GLOB)")
file_paths = glob.glob('/content/sample_data/sales*.csv')
sales_df_multi = pd.concat(
    [pd.read_csv(fp) for fp in file_paths],
    ignore_index=True
)
print(f"Multi-file shape: {sales_df_multi.shape}")
display(sales_df_multi.head())

# ===== 3. EXTRACT: SELECTIVE LOOP (FIRST 5 DAYS) =====
print("\nðŸ“¥ 3. EXTRACT SELECTIVE FILES (LOOP)")
sales_df_loop = pd.DataFrame()
# Use the file_paths generated in section 2 and take the first 5
for i, filepath in enumerate(file_paths[:5]): # Take first 5 files from the glob list
    day_df = pd.read_csv(filepath)
    sales_df_loop = pd.concat([sales_df_loop, day_df], ignore_index=True)
    print(f'File {i+1}: {len(sales_df_loop)} rows (from {filepath})')
print(f"Loop shape: {sales_df_loop.shape}")
display(sales_df_loop.head())

# ===== 4. EXTRACT: JSON & PARQUET =====
print("\nðŸ“¥ 4. EXTRACT JSON/PARQUET")
sales_json = pd.read_json('/content/sample_data/LA_Retail_Sales.json', lines=True)
print(f"JSON shape: {sales_json.shape}")

sales_parquet = pd.read_parquet('/content/sample_data/LA_Retail_Sales.parquet')
print(f"Parquet shape: {sales_parquet.shape}")

# ===== 5. EXTRACT: DATABASE (CREATE SAMPLE TABLE FIRST) =====
print("\nðŸ“¥ 5. EXTRACT FROM DATABASE")
# Create sample sales table for demo
sample_sales = sales_df.head(100).copy()
# Standardize column names before writing to SQL
sample_sales.columns = sample_sales.columns.str.lower().str.replace(' ', '_')
sample_sales.to_sql('sales', con, if_exists='replace', index=False);

db_sales = pd.read_sql('SELECT * FROM sales', con=con)
print(f"DB extract shape: {db_sales.shape}")

db_agg = pd.read_sql("""
    SELECT store_id, SUM(dollar_sales) AS total_sales
    FROM sales GROUP BY store_id ORDER BY total_sales DESC
""", con=con)
db_agg.plot(kind='bar', x='store_id', y='total_sales')
plt.title('Sales by Store')
plt.show()

# ===== 6. EXTRACT: API (OPENWEATHER - ENRICH WITH LA WEATHER) =====
print("\nðŸ“¥ 6. API EXTRACT (Weather Enrichment)")
# âš ï¸ REPLACE WITH YOUR API KEY (free tier limited; historical needs paid)
API_KEY = '4b8451b998e260dde3264af310ebe944'  # https://openweathermap.org/api
lat, lon = 34.0522, -118.2437  # Los Angeles

# Use sample dates from sales data
dates = pd.date_range('2024-09-01', '2024-09-30', freq='D').strftime('%Y-%m-%d').tolist()
print(f"Fetching weather for {len(dates)} dates...")

weather_data = []
for date_str in dates[:5]:  # Limit to 5 for free tier demo
    unix_ts = int(datetime.strptime(date_str, '%Y-%m-%d').timestamp())
    url = f'https://api.openweathermap.org/data/3.0/onecall/timemachine?lat={lat}&lon={lon}&dt={unix_ts}&appid={API_KEY}&units=imperial'
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            if 'data' in data and len(data['data']) > 0:
                w = data['data'][-1]
                weather_data.append({
                    'date': date_str,
                    'temp_f': w.get('temp', 0),
                    'humidity': w.get('humidity', 0),
                    'description': w['weather'][0]['description']
                })
        time.sleep(1)  # Rate limit
    except Exception as e:
        print(f"API error for {date_str}: {e}")

weather_df = pd.DataFrame(weather_data)

# Handle case where weather_data might be empty due to API key or other issues
if not weather_df.empty:
    weather_df['date'] = pd.to_datetime(weather_df['date'])
    print(f"Weather shape: {weather_df.shape}")
    display(weather_df.head())
    weather_df.set_index('date').plot()
    plt.title('Weather Trends')
    plt.show()
else:
    print("No weather data retrieved. Skipping weather trend plot.")

# ===== 7. TRANSFORM (USE MAIN SALES_DF) =====
print("\nðŸ”„ 7. TRANSFORM")
sales_final = sales_df.copy()  # Work on main dataset

# Standardize columns
sales_final.columns = sales_final.columns.str.lower().str.replace(' ', '_')

# Handle missing (drop for demo)
sales_final = sales_final.dropna(subset=['unit_sales', 'dollar_sales'])
print(f"After dropna: {sales_final.shape}")

# Engineer features
sales_final['revenue_per_unit'] = (sales_final['dollar_sales'] / sales_final['unit_sales']).round(2)

# Fix strings
sales_final['store_zip'] = sales_final['store_zip'].astype(str).str.replace('xx', '00', regex=False)

# Cast types
sales_final['date'] = pd.to_datetime(sales_final['date'])
sales_final['unit_sales'] = sales_final['unit_sales'].astype(int)
sales_final['promotion_flag'] = sales_final['promotion_flag'].astype(bool)

# Merge weather
# Only merge if weather_df is not empty
if not weather_df.empty:
    weather_merge = weather_df[['date', 'temp_f']].copy()
    weather_merge.columns = ['date', 'temp']
    sales_final = sales_final.merge(weather_merge, on='date', how='left')
else:
    print("No weather data to merge.")
    sales_final['temp'] = None # Add a temp column with nulls if no weather data

# QA
print(sales_final.describe())
sns.histplot(data=sales_final, x='dollar_sales', kde=True)
plt.title('Dollar Sales Distribution')
plt.show()

display(sales_final.head())
sales_final.info()

# ===== 8. LOAD TO DATABASE =====
print("\nðŸ’¾ 8. LOAD")
sales_final.to_sql('sales_final_etl', con, if_exists='replace', index=False)
print(f"âœ… Loaded {len(sales_final)} rows to 'sales_final_etl' table")

# Verify
loaded = pd.read_sql('SELECT COUNT(*) as rows, AVG(temp) as avg_temp FROM sales_final_etl', con=con)
display(loaded)
hot_days = pd.read_sql('SELECT * FROM sales_final_etl WHERE temp > 70 LIMIT 5', con=con)
display(hot_days)

print("\nðŸŽ‰ ETL PIPELINE COMPLETE!")
print("ðŸ“Š Run this query for analysis: SELECT * FROM sales_final_etl WHERE promotion_flag = 1")
print("ðŸ”„ To automate: Wrap in function + Airflow/AWS Lambda for scheduling")
```
