```python
# SciPy Tutorial: Complete Python Script
# Run in Jupyter or as .py file (pip install scipy numpy matplotlib)
import numpy as np
import scipy
from scipy import stats, integrate, optimize, linalg
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

print("SciPy version:", scipy.__version__)
print("=" * 50)

# 1. Basic Statistics Module
print("\n1. STATISTICS MODULE")
data = np.array([1, 2, 3, 4, 5])
result = stats.describe(data)
print("Dataset:", data)
print("Summary stats:", result)
print("Trimmed mean (limits=(2,4)):", stats.tmean(data, (2, 4)))

# 2. Integration Example
print("\n2. INTEGRATION")
def integrand(x):
    return np.sin(x)
integral, error = integrate.quad(integrand, 0, np.pi)
print("Integral of sin(x) from 0 to pi:", integral, "Error:", error)

# 3. Linear Algebra
print("\n3. LINEAR ALGEBRA")
A = np.array([[2, 1], [4, -5]])
b = np.array([[5], [7]])
x = linalg.solve(A, b)
print("Matrix A:\n", A)
print("Vector b:\n", b)
print("Solution x:\n", x)
print("Verification (Ax - b):\n", np.dot(A, x) - b)

# 4. Optimization
print("\n4. OPTIMIZATION")
def rosen(x):
    return np.sum(100.0 * (x[1:] - x[:-1]**2.0)**2.0 + (1 - x[:-1])**2.0)
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = optimize.minimize(rosen, x0, method='BFGS')
print("Rosenbrock minimum at:", res.x)
print("Function value:", res.fun)

# 5. Interpolation and Plot
print("\n5. INTERPOLATION")
x = np.linspace(0, 10, 10)
y = np.sin(x) + np.random.normal(0, 0.1, 10)
f = interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 100)
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'o', label='Original data')
plt.plot(x_new, f(x_new), '-', label='Cubic interpolation')
plt.title('SciPy Interpolation Example')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.savefig('scipy_interp.png')  # Saves plot to file
plt.show()
print("Plot saved as 'scipy_interp.png'")

print("\nTutorial complete! Explore subpackages like scipy.signal or scipy.special next.")
```
