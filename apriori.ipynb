```python
# Import packages
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Create data
data = [['Apple', 'Banana', 'Guava', 'Eggs', 'Milk', 'Steak', 'Yogurt', 'Tomatos', 'Onion', 'Nuts'],
        ['Apple', 'Banana', 'Orange', 'Eggs', 'Milk', 'Pork', 'Potatos', 'Onion', 'Choclate'],
        ['Spinach', 'Eggs', 'Milk', 'Yogurt', 'Bagel', 'Onion', 'Watermelon'],
        ['Pineapple', 'Banana', 'Eggs', 'Lemon', 'Steak', 'Broccolli', 'Onion', 'Lamb'],
        ['Apple', 'Banana', 'Beer', 'Lettuce', 'Ice cream', 'Tomatos', 'Grape', 'Strawberry']]

# One-hot encoding
te = TransactionEncoder()
data_transformed = te.fit_transform(data)
df = pd.DataFrame(data_transformed, columns=te.columns_)
df

# Find frequent item sets
frequent_items = apriori(df, min_support = 0.6, use_colnames = True)

# Generate strong association rules
rules = association_rules(frequent_items, metric ="confidence", min_threshold = 0.7)
rules = rules.sort_values(by='confidence', ascending =False)
rules
```

```python
rules1=rules[['antecedents','consequents','support','confidence','lift']]
rules1
```

```python
rules2=rules1[rules1['confidence']>=1.0]
rules2
```
