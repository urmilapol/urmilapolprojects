```python
# Complete Kafka setup and demo in one cell
!apt-get update && apt-get install -y openjdk-11-jdk
!wget https://downloads.apache.org/kafka/3.7.0/kafka_2.13-3.7.0.tgz && tar -xzf kafka_2.13-3.7.0.tgz

# Start services
!cd kafka_2.13-3.7.0 && ./bin/zookeeper-server-start.sh -daemon config/zookeeper.properties
!cd kafka_2.13-3.7.0 && ./bin/kafka-server-start.sh -daemon config/server.properties

import time
time.sleep(15) # Increased sleep time to give Kafka services more time to start

!pip install kafka-python

# Create topic
!cd kafka_2.13-3.7.0 && ./bin/kafka-topics.sh --create --topic orders_topic --bootstrap-server localhost:9092 --partitions 1 --replication-factor 1

# Import and setup
from kafka import KafkaProducer, KafkaConsumer
import json
import random

# Producer
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Send test data
for i in range(3):
    order = {'quantity': random.randint(1, 5), 'price': round(random.uniform(20, 50), 2)}
    producer.send('orders_topic', order)
    print(f"Produced: {order}")
producer.flush()

# Consumer
consumer = KafkaConsumer(
    'orders_topic',
    bootstrap_servers='localhost:9092',
    value_deserializer=lambda m: json.loads(m.decode('utf-8')),
    auto_offset_reset='earliest'
)

print("\nConsuming messages...")
for message in consumer:
    order = message.value
    total = order['quantity'] * order['price']
    print(f"Order Total: ${total:.2f}")
    break  # Just show first message for demo

# Cleanup
consumer.close()
```
