# Comprehensive Data Analysis Pipeline with Pandas
**Prepared by:** Dr. Urmila Pol, Associate Professor, Department of Computer Science Shivaji University Kolhapur

This notebook contains the complete, step-by-step implementation of basic to advanced workflows using the Python Pandas library, covering:
1. Environment Setup & Data Creation
2. Basic Profiling & High-Level Inspection
3. Data Selection, Slicing & Advanced Filtering (`loc`, `iloc`, `query`)
4. Data Transformation (Adding, Updating, and Dropping Rows/Columns)
5. Working with Date/Time Objects
6. Handling Missing Data (Imputation)
7. Grouping, Aggregation & Data Merging

```python
# ==========================================
# 1. SETUP AND CORE DATAFRAME CREATION
# ==========================================
print("=== 1. Setup & Data Creation ===")

import pandas as pd
import numpy as np

# Print the dynamic library version
print(f"Pandas Version: {pd.__version__}\n")

# Initial dictionary dataset matching the tutorial
data = {
    'Name': ['Madhav', 'Vishakha', 'Lalita', 'Rishabh'],
    'Age': [16, 17, 18, 19],
    'Salary': [90000, 70000, 50000, 30000]
}

# Instantiate DataFrame and apply professional column headers
df = pd.DataFrame(data)
df.rename(columns={'Salary': 'Monthly Salary'}, inplace=True)
print("Initial Structural DataFrame:")
print(df)
print("-" * 50)


# ==========================================
# 2. BASIC PROFILE INSPECTION
# ==========================================
print("\n=== 2. Basic Profile Inspection ===")

print(f"DataFrame Dimensional Shape (Rows, Columns): {df.shape}")
print("\nTop 2 Rows (df.head):")
print(df.head(2))

print("\nBottom 2 Rows (df.tail):")
print(df.tail(2))

print("\nHigh-Level Descriptive Statistical Summary:")
print(df.describe())
print("-" * 50)


# ==========================================
# 3. TRANSFORMATION: ADDING, UPDATING & DROPPING
# ==========================================
print("\n=== 3. Data Transformation ===")

# Adding Static and Computed Scalar Fields
df['Team'] = ['CEO', 'HR', 'CTO', 'Data Analyst']
df['Bonus'] = df['Monthly Salary'] * 0.20  # 20% bonus calculation

# Appending a structural row dynamically using length property
new_row = ['ABC', 20, 45000, 'Marketing', 9000]
df.loc[len(df)] = new_row
print("DataFrame After Structural Additions:")
print(df)

# Updating explicit values via index labels
df.loc[0, 'Monthly Salary'] = 95000  # Changing Madhav's salary

# Dropping redundant metadata (dropping row 'ABC' and column 'Bonus')
df.drop(index=4, inplace=True)
df.drop(columns=['Bonus'], inplace=True)
print("\nDataFrame After In-Place Specific Cleanups:")
print(df)
print("-" * 50)


# ==========================================
# 4. DATA SELECTION AND FILTERING METHODS
# ==========================================
print("\n=== 4. Advanced Selection & Filtering ===")

# loc: Label-based filtering with boolean masking
print("Label-Based Selection (loc) - Filtered rows where Age >= 17:")
print(df.loc[df['Age'] >= 17])

# iloc: Positional-based indexing (Slicing the first 2 rows completely)
print("\nPositional-Based Selection (iloc) - First 2 rows:")
print(df.iloc[0:2])

# Query Method: Clean, declarative SQL-like string querying
print("\nSQL-Like Query Selection - Age >= 17 and Monthly Salary > 40000:")
print(df.query("Age >= 17 and `Monthly Salary` > 40000"))
print("-" * 50)


# ==========================================
# 5. WORKING WITH DATE & TIME DATA TYPES
# ==========================================
print("\n=== 5. Working with Date & Time Data Types ===")

# Setting distinct date formatting variations
df['Date of Joining'] = ['2024-01-01', '2024-01-15', '2024-03-10', '2024-03-28']

# Type-casting raw Object vectors into system DateTime formats
df['Date of Joining'] = pd.to_datetime(df['Date of Joining'])

# Extracting temporal component values
df['Month'] = df['Date of Joining'].dt.month
df['Day Name'] = df['Date of Joining'].dt.day_name()

# Applying TimeDelta scalar offsets to calculate projection periods
df['Probation End Date'] = df['Date of Joining'] + pd.to_timedelta(90, unit='d')

print("DataFrame Optimized with Structured Temporal Metrics:")
print(df[['Name', 'Date of Joining', 'Month', 'Day Name', 'Probation End Date']])
print("-" * 50)


# ==========================================
# 6. HANDLING MISSING VALUES (IMPUTATION)
# ==========================================
print("\n=== 6. Handling Missing Values ===")

# Force-injecting a missing entity value (NaN) into target row vector
df.loc[3, 'Monthly Salary'] = np.nan
print("Missing Value Breakdown Per Axis Column Vector:")
print(df.isnull().sum())

# Implementing imputation values to maintain dataset integrity
df['Monthly Salary'] = df['Monthly Salary'].fillna(0)
print("\nDataFrame Post-Imputation Setup (NaN Replaced with 0):")
print(df[['Name', 'Monthly Salary']])
print("-" * 50)


# ==========================================
# 7. AGGREGATION, GROUPBY & MERGING
# ==========================================
print("\n=== 7. Aggregation, GroupBy & Merging ===")

# Multiple Aggregation parsing conditional targets
print("Aggregated Monthly Performance Matrix (Mean Salary and Name Count by Month):")
grouped_summary = df.groupby('Month').agg({
    'Monthly Salary': 'mean',
    'Name': 'count'
})
print(grouped_summary)

# Setup structural keys to mimic operational Join integrations
df_left = pd.DataFrame({'ID': [1, 2, 3], 'Employee_Name': ['Madhav', 'Vishakha', 'Lalita']})
df_right = pd.DataFrame({'ID': [1, 2, 4], 'Performance_Score': [95, 88, 72]})

# Executing horizontal join strategies via explicit column indexing keys
df_merged = pd.merge(df_left, df_right, on='ID', how='inner')
print("\nResultant Inner Join Integration Output via Common ID Keys:")
print(df_merged)
print("\n" + "="*40 + "\nExecution Pipeline Finished Successfully.\n" + "="*40)
```
