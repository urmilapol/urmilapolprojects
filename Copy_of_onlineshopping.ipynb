```python
import pandas as pd

# -----------------------------
# STEP 1: Data Ingestion
# -----------------------------
print("Loading order data...")
orders = pd.read_csv("/content/sample_data/orders.csv")
print(orders)

# -----------------------------
# STEP 2: Data Cleaning
# -----------------------------
print("\nCleaning data...")
orders = orders.dropna()              # Remove missing values
orders['quantity'] = orders['quantity'].astype(int)
orders['price'] = orders['price'].astype(float)

# -----------------------------
# STEP 3: Data Transformation
# -----------------------------
print("\nTransforming data...")
orders['total_amount'] = orders['quantity'] * orders['price']

# Sales by category
category_sales = orders.groupby('category')['total_amount'].sum().reset_index()

# Sales by product
product_sales = orders.groupby('product')['total_amount'].sum().reset_index()

# -----------------------------
# STEP 4: Data Storage
# -----------------------------
orders.to_csv("clean_orders.csv", index=False)
category_sales.to_csv("category_sales_report.csv", index=False)
product_sales.to_csv("product_sales_report.csv", index=False)

# -----------------------------
# STEP 5: Output
# -----------------------------
print("\nClean Orders Data:")
print(orders)

print("\nCategory-wise Sales Report:")
print(category_sales)

print("\nProduct-wise Sales Report:")
print(product_sales)

print("\nData Pipeline Executed Successfully!")
```
